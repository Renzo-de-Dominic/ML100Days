```python
import numpy as np
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 新建地圖
map = Basemap(projection = 'ortho', lat_0 = 0, lon_0 = 0)


# 畫圖
map.drawcoastlines()

# 顯示结果
plt.show()

# 存储结果
plt.savefig('ortho_earth.png')
```

```python
# 新建地圖
map2 = Basemap(projection = 'mill')


# 畫圖
map2.drawcoastlines()

# 顯示结果
plt.show()

# 存储结果
plt.savefig('Mill_earth.png')
```

```python
# 運用 "itertools" 庫中的 "chain" 模組，可以繪製更接近於生活中所見世界地圖的圖像
from itertools import chain 
fig = plt.figure(figsize=(8, 6), edgecolor='w')
Map = Basemap(projection='cyl', resolution='f',
            llcrnrlat=22, urcrnrlat=25,
           llcrnrlon=120, urcrnrlon=122, )

# 繪製一個陰影浮雕的地理圖像
Map.shadedrelief(scale=0.2)


```

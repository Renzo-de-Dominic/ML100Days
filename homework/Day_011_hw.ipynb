```python
import numpy as np
import pandas as pd
```

```python
data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data, index=labels)
df
```

```python
ddf=df.sort_values(by=['animal'])
ddf
```

```python
#計算每個不同種類 animal 的 age 的平均數
catdata=ddf[0:4]
print('cat mean age',catdata.age.mean())
dogdata=ddf[4:8]
print('dog mean age',dogdata.age.mean())
snakedata=ddf[8:10]
print('snake mean age',snakedata.age.mean())
```

```python
#將資料依照 Age 欄位由小到大排序，再依照 visits 欄位由大到小排序
print(df.sort_values(by=['age']))
print(df.sort_values(by=['visits'], ascending=False))
```

```python
#將 priority 欄位中的 yes 和 no 字串，換成是布林值 的 True 和 False
dtf=df.replace('yes','True')
dvf=dtf.replace('no','False')
dvf
```

```python
#3.一個包含兩個欄位的 DataFrame#，哪一筆的資料總合最小
df4=df3.sum()
#不漏掉任一筆最小值:
df4[df4 == df4.min()].dropna()
```
